```python
article_text = "This is the text you want to clean. It may contain spaces,  punctuations, and numbers."
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
clean_text = re.sub('[^a-zA-Z]', ' ', article_text)
print(clean_text)
```

```python
article_text = article_text.lower()
print(article_text)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Download the punkt_tab resource
sentence_list = nltk.sent_tokenize(article_text)
print(sentence_list)
```

```python
nltk.download('punkt_tab') # Download the punkt_tab resource
tokens = word_tokenize(article_text)
print(tokens)
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
word_frequencies = {}
if 'tokens' in locals(): # Check if tokens is defined before iterating
    for word in tokens:
        if word not in stopwords:
            if word not in word_frequencies:
                word_frequencies[word] = 1
            else:
                word_frequencies[word] += 1
print(word_frequencies)
```

```python
import string
from nltk.probability import FreqDist
remove_these = set(stopwords + list(string.punctuation) + list(string.digits))
word_freq = FreqDist(word_frequencies)
print(word_freq)
```

```python
word_freq.plot(10, title="freqplot")
```

```python
from collections import Counter
dictionary_counter = Counter(word_frequencies)
print(dictionary_counter)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter # Import Counter

if 'word_frequencies' in locals(): # Check if word_frequencies is defined
    dictionary_counter = Counter(word_frequencies) # Define dictionary_counter
    cloud = WordCloud(max_font_size=80, colormap='hsv').generate_from_frequencies(dictionary_counter)
    plt.figure(figsize=(16,12))
    plt.imshow(cloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
else:
    print("word_frequencies is not defined.")
```
